```python
import pandas as pd
import numpy as np
import timeit

file_path = 'data/household_power_consumption.txt'

def load_and_clean(path):
    df = pd.read_csv(path, sep=';', low_memory=False)
    
    df.replace('?', np.nan, inplace=True)
    df.dropna(inplace=True)
    
    numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 
                    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col])
        
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
    return df

df = load_and_clean(file_path)
print(f"Завантажено записів: {len(df)}")
display(df.head())
```

```python
def task_1(df):
    return df[df['Global_active_power'] > 5]

t1_time = timeit.timeit(lambda: task_1(df), number=1)
print(f"Час виконання: {t1_time:.5f} сек")
display(task_1(df).head())
```

```python
def task_2(df):
    return df[(df['Global_intensity'] >= 19) & 
              (df['Global_intensity'] <= 20) & 
              (df['Sub_metering_2'] > df['Sub_metering_3'])]

t2_time = timeit.timeit(lambda: task_2(df), number=1)
print(f"Час виконання: {t2_time:.5f} сек")
display(task_2(df).head())
```

```python
def task_3(df):
    sample = df.sample(n=500000, replace=False) 
    return sample[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()

t3_time = timeit.timeit(lambda: task_3(df), number=1)
print(f"Час виконання: {t3_time:.5f} сек")
print("Середні значення груп споживання:")
print(task_3(df))
```

```python
import datetime

def task_4(df):
    time_limit = datetime.time(18, 0, 0)
    subset = df[(df['Time'] > time_limit) & (df['Global_active_power'] > 6)]
     
    subset = subset[(subset['Sub_metering_2'] > subset['Sub_metering_1']) & 
                    (subset['Sub_metering_2'] > subset['Sub_metering_3'])]
     
    half = len(subset) // 2
    first_half = subset.iloc[:half:3] 
    second_half = subset.iloc[half::4]
    
    return pd.concat([first_half, second_half])

t4_time = timeit.timeit(lambda: task_4(df), number=1)
print(f"Час виконання: {t4_time:.5f} сек")
display(task_4(df).head())
```
